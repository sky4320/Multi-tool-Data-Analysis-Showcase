# Task 2: Aggregation & Filtering & Rank
In this task, we will perform some very high level aggregation and filtering operations. 
Then, we will apply ranking on the results for some tasks. 
Pandas has a convenient and powerful syntax for aggregation, filtering, and ranking. 
DO NOT write a for loop. Pandas has built-in functions for all tasks. 

```python
# 20 points
# Task 2-A: Generate a Outside_Temperature.txt file containing the answers to the following questions: 
# Using the “Outside Temperature” values: 
# a. What is the average time of hottest daily temperature (over month);
#creates a file in append if it is a new file or opens an existing file
f = open("Outside_Temperature.txt", "a")
f.write('Average time of hottest daily temperature\n\n')

#grouped the data by date and outside temperature
datagrouped = dataset.groupby('Date')['Outside Temperature']

#Here we are getting the max temp and if there are mutiple max values, it is multipled by 10 as we have 10 min time difference in dataset.
hottestdaily = (datagrouped.agg(lambda v: v.value_counts().loc[v.max()]*10)
   .reset_index(name='Averagetime')
)
f.write(hottestdaily.to_string())

f.write('\n Average time of hottest daily temperature (over month) :')

#Calculating the average time for month.
f.write(str(sum(hottestdaily.to_dict("list")['Averagetime'][1:31])/60))


# b. What time of the day is the most commonly occurring hottest time; 

f.write( '\n Most commonly occurring hottest time \n')

#creating a temporary max column which containes the max temprature for that date.
dataset['Max'] = datagrouped.transform(max)

data = pd.Series(dataset.loc[dataset['Outside Temperature'] == dataset['Max']]['Time']) 
f.write(data.value_counts().to_string())
f.write(' \n\n Time of the day is the most commonly occurring hottest time 11:00 TO 15:00')


# c. Which are the Top Ten hottest times on distinct days, prefera
f.write( '\n\n Top Ten hottest times \n\n\n')

#Here we are getting the max temp.
data = (datagrouped
   .agg(lambda v: v.max()) # each recording is for 10 minutes period
   .reset_index(name='hightemp')
)
f.write(str(data.sort_values('hightemp',ascending = False).head(10).sort_values('Date',ascending=True))) #Getting top 10 records
f.close()
```
