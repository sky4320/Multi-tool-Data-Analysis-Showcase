```r
# Task 1: Statistical Exploratory Data Analysis
```

```r
#1-a Print the details of dataframe
str(hcs_data)  # displaying internal structure of the data frame

cat("\n checking of data which is missing")

hcs_data <- hcs_data %>% mutate(smoking_status = replace(smoking_status, smoking_status ==  "-", NA)) 

cbind(lapply(lapply(hcs_data, is.na), sum)) 
```

```r
#1-b Find the number of rows and columns in dataset
rows<-nrow(hcs_data)           # assigning rows variable with the number of rows of the data set using nrow() fucntion
cols<-ncol(hcs_data)           # assigning columns variable with the number of columns of the data set using nrow() fucntion
cat("Total rows in the dataset = ", rows) #Printing number of rows in the dataset
cat("\nTotal columns in the dataset = ", cols) #Printing number of columns in the dataset
```

```r
#1-c Print descriptive detail of a column in dataset
count<- nrow(hcs_data[hcs_data$heart_disease]) # Assigning count variable with heartdisease row by using nrow() function
unique_val<- unique(hcs_data$heart_disease)    # Retrieving unique value from heart disease row by using unique() function
freq_val<- mfv(hcs_data$heart_disease)         # Retrieving most frequent values from heart disease row using mfv() function
top_value<- max(hcs_data$heart_disease)        # Retrieivng maximum value from heart disease row using max() function

cat("Heart disease column details\n")

cat("\nThe count is:",count)                         #Printing number of values in heart disease column
cat("\nThe unique Values are:",unique_val)            #Printing unique values present in heart disease column
cat("\nThe frequent values are:",freq_val)            #Printing frequent values present in heart disease column
cat("\nThe top value is:",top_value)                 #Printing the maximum value from the heart disease column
```

```r
#1-d Find all the count of unique values for a 'avg_glucose_level' column in dataset
countof_unqvalues_glucose <-unique(hcs_data$avg_glucose_level) #Retrieving unique value from avg_glucose_level column
length(countof_unqvalues_glucose)

```

```r
#1-d Find all percentage of 'Residence_type' for all the values
count <- nrow(hcs_data[hcs_data$Residence_type])

urban <- nrow(subset(hcs_data,Residence_type=="Urban"))
rural <- nrow(subset(hcs_data,Residence_type=="Rural"))


cat("\nUrban residence type percentage:",(urban/count)*100)
cat("\nRural residence type percentage:",(rural/count)*100)
```

```r
# Task 2: Aggregation & Filtering & Rank
```

```r
#Task 2-a: Find out the gender with largest number of records
numoffemales <- nrow(subset(hcs_data,gender=="Female"))  #retrieving female subset from the dataset
numofmales <- nrow(subset(hcs_data,gender=="Male"))      #retrieving male subset from the dataset
otherscount <- nrow(subset(hcs_data,gender=="Other"))    #retrieving other subset from the dataset

if(numoffemales>numofmales & otherscount<numoffemales) {           #conditions to find the highest count in the gender
    {
        print("Record is larger in Females")
    }
} else if(otherscount<numofmales){
    print("Record is larger in Males")
} else
    print("Record is larger in Others")
```

```r
#Task 2-b: Find out the total number of Residence_type "Urban" who are Male

totalResType <- subset(hcs_data,Residence_type=="Urban" & gender=="Male")   #retreiving the gender male and who has residence type as urban from the data set
cat("The total number of Residence_type Urban who are Male:", nrow(totalResType)) #printing total number of males who have the urban residency status
```

```r
# Group by function for dataframe in R using pipe operator 
#2-c 1 question #Find the top 10 ages with highest av_glucose_level
cat("\n The top 10 Ages with highest average glucose levels:")
top10 <-select(hcs_data, age, avg_glucose_level)
hcs_data %>%
group_by(age) %>%
summarize(
            avg_glucose_level = max(avg_glucose_level, na.rm = TRUE)) %>%
arrange(desc(avg_glucose_level)) %>%
head(10)
```

```r
#2-d 2nd question top 10 ages with more number of strokes
cat("\n Top 10 Ages with more number of strokes")
top10 <-select(hcs_data, age, stroke)
hcs_data %>%
group_by(age) %>%
summarize(
            no.of.strokes = sum(stroke, na.rm = TRUE)) %>%
arrange(desc(no.of.strokes)) %>%
head(10)
```

```r
#task 3-a
#Create barplot showing gender with count with residence type
restype <- c("Urban","Rural")        #creating residence type vector for urban and rural restype
dataset <- read.csv("healthcare_stroke_dataset.csv") #reading the hcs dataset
gender <- c("Male","Female","Other")    #creating genders vector
colors= c("red","blue","green")         #creating color vectors
Maleagedata <- c()
Femaleagedata <- c()
Otheragedata <- c()

for (i in gender)
    {
    Urban <- nrow(subset(dataset, (Residence_type=="Urban") & gender==i))
    Rural <- nrow(subset(dataset, (Residence_type=="Rural") & gender==i))
    if(i=="Male")
        {
        Maleagedata <- c(Urban,Rural)
    }else if(i=="Female")
        {
        Femaleagedata <- c(Urban,Rural) 
    }else
        {
        Otheragedata <- c(Urban,Rural)
    }
        
}

Finalplot <- rbind(Maleagedata,Femaleagedata,Otheragedata)

barplot(Finalplot,beside=T,names.arg=restype,xlab="Residence Type",ylab="Gender Count",col=colors)
legend("bottomright", gender, cex = 1, fill = colors)
```

```r
#task 3-b
#Display pie chart for the smoking status data
dataset <- read.csv("healthcare_stroke_dataset.csv")
colors <- c("red","blue","green")
smokestatus <- c("formerly smoked","never smoked","smokes")
smokestatusdata <- c()
for (i in smokestatus)
    {
    number <- nrow(subset(dataset,smokestatus==i))
    smokestatusdata <- c(smokestatusdata,number)
}
smokepercent <- round(smokestatusdata*100/sum(smokestatusdata),1)
pie(smokestatusdata,labels=smokepercent,main="Smoke type chart",col=colors)
legend("bottomleft", smokestatus, cex = 0.9, fill=colors)
```
